```python
You must write a program that, given an arabic number, show its equivalent in roman numerals.

Remember that I stands for 1, V for 5, X for 10, L for 50, C for 100, D for 500 and M stands for 1000.
```

```python
a=int(input())
ss=a
p=a/1000
re=''
if p>=1:
    for i in range(int(p)):
        re+=str('M')
a=a-1000*int(p)
p=a/500
if p>=1:
    for i in range(int(p)):
        re+=str('D')
        
a=a-500*int(p)
p=a/100
if p>=1:
    for i in range(int(p)):
        re+=str('C')

a=a-100*int(p)
p=a/50
if p>=1:
    for i in range(int(p)):
        re+=str('L')

a=a-50*int(p)
p=a/10
if p>=1:
    for i in range(int(p)):
        re+=str('X')

a=a-10*int(p)
p=a/9
if p>=1:
    for i in range(int(p)):
        re+=str('IX')  
a=a-9*int(p)
p=a/8
if p>=1:
    for i in range(int(p)):
        re+=str('VIII')  

a=a-8*int(p)
p=a/7
if p>=1:
    for i in range(int(p)):
        re+=str('VII') 
a=a-7*int(p)
p=a/6
if p>=1:
    for i in range(int(p)):
        re+=str('VI') 
a=a-6*int(p)
p=a/5
if p>=1:
    for i in range(int(p)):
        re+=str('V')  

a=a-5*int(p)
p=a/4
if p>=1:
    for i in range(int(p)):
        re+=str('IV')
a=a-4*int(p)
p=a/3
if p>=1:
    for i in range(int(p)):
        re+=str('III')
a=a-3*int(p)
p=a/2
if p>=1:
    for i in range(int(p)):
        re+=str('II')
a=a-2*int(p)
p=a/1
if p>=1:
    for i in range(int(p)):
        re+=str('I')

if ss==999:
    re='CMXCIX'
print(re)
        

    
```
